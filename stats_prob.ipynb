```python
# remove all rows that have NaN
# boxplot cannot be made for a dataframe having NaN values
df2 = df.dropna()
df2
```

```python
plt.hist(df['eDPI'])
plt.show()
```

```python
#figure should be 10 inch x 2 inch
plt.figure(figsize=(10,2))
# means is green triangle
plt.boxplot(df2['eDPI'], vert=False, showmeans=True)
plt.grid(color='gray', linestyle='dotted')
plt.tight_layout()
plt.show()
# orange line is the median
```

```python
df.boxplot(column='eDPI', by='Grip', figsize=(10,8))
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.show()
```
