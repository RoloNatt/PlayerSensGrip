```python
# Shapiro-Wilk test
from scipy.stats import shapiro


# perform Shapiro-Wilk test
stat, p = shapiro(df['eDPI'])

# we can accept 5% chance of making a Type I error 
# lowering the significance level increases the risk of a Type II error, which occurs when we fail to reject the null hypothesis when it is actually false. 
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)') # Gaussian = Normal distribution
else:
    print('Sample does not look Gaussian (reject H0)')
```

```python
# Confidence Intervals
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    # convert the data input into a NumPy array (for older python versions - By multiplying the array by a floating point number 1.0, we ensure that the resulting array will also be of floating point type. If we didn't do this, the result of the division in the mean calculation would be truncated to an integer, which could lead to incorrect results.)
    a = 1.0 * np.array(data)
    n = len(a)
    # calculate the mean and standard error of the mean for 'a' 
    m, se = np.mean(a), scipy.stats.sem(a)
    # calculate the half-width of the confidence interval using the t-distribution with the specified confidence level and degrees of freedom (n-1)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, h

for p in [0.85, 0.9, 0.95]:
    m, h = mean_confidence_interval(df['eDPI'].fillna(method='pad'),p)  # missing values filled using the previous value
    print(f"p={p:.2f}, mean = {m:.2f} ± {h:.2f}")
```

```python
# Hypothesis Testing
# agg = aggregating = summarizing
# You can use any column name in the agg function for counting, as long as that column contains non-null values for all the rows in each group.
df.groupby('Grip').agg({ 'DPI' : 'mean', 'eDPI' : 'mean', 'Playstyle' : 'count'}).rename(columns={ 'Playstyle' : 'Count'})
```

```python
# test the confidence intervals
for p in [0.85,0.9,0.95]:
    m1, h1 = mean_confidence_interval(df.loc[df['Grip']=='claw',['eDPI']],p)
    m2, h2 = mean_confidence_interval(df.loc[df['Grip']=='palm',['eDPI']],p)
    print(f'Conf={p:.2f}, claw: {m1-h1[0]:.2f}..{m1+h1[0]:.2f}, palm: {m2-h2[0]:.2f}..{m2+h2[0]:.2f}')

# the intervals are overlapping, so we cannot say for sure that players with palm grip have more eDPI than players with claw grip    
```

```python
for p in [0.85,0.9,0.95]:
    m1, h1 = mean_confidence_interval(df2.loc[df2['Playstyle']=='aim',['eDPI']],p)
    m2, h2 = mean_confidence_interval(df2.loc[df2['Playstyle']=='movement',['eDPI']],p)
    print(f'Conf={p:.2f}, aim: {m1-h1[0]:.2f}..{m1+h1[0]:.2f}, movement: {m2-h2[0]:.2f}..{m2+h2[0]:.2f}')
```

```python
# Student t-test - statistically more correct way to prove the hypothesis
# ttest_ind = t-test independent
from scipy.stats import ttest_ind
# equal_var=False --> test should not assume equal variances between the two groups
tval, pval = ttest_ind(df.loc[df['Grip']=='claw',['eDPI']], df.loc[df['Grip']=='palm',['eDPI']],equal_var=False)
print(f"T-value = {tval[0]:.2f}\nP-value: {pval[0]}")
```

```python
# Simulating a Normal Distribution with the Central Limit Theorem
def normal_random(sample_size=100):
    sample = [random.uniform(0,1) for _ in range(sample_size) ]
    return sum(sample)/sample_size

sample = [normal_random() for _ in range(100)]
plt.figure(figsize=(10,6))
plt.hist(sample)
plt.tight_layout()
plt.show()
```
